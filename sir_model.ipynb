```python
import numpy as np 
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from evolve import evolve, show, print_args

# Derivatives definition
def S_dot_f(state, beta):
    return - beta * state[0] * state[1]

def I_dot_f(state, alpha, beta):
    return beta * state[0] * state[1] - alpha * state[1]

def R_dot_f(state, alpha):
    return alpha * state[1]

def deriv(state, alpha, beta):
    S_dot = S_dot_f(state, beta)
    I_dot = I_dot_f(state, alpha, beta)
    R_dot = R_dot_f(state, alpha)
    N_dot = 0
    return np.asarray([S_dot, I_dot, R_dot, N_dot])
```

```python
# initializations
N0 = 100
S0 = 95
I0 = 5
R0 = 0
state = np.asarray([S0, I0, R0, N0])

timelength = 1000
kwargs = {'alpha' : 0.003, 'beta' : 0.002}
legend = ['susceptibles', 'infected', 'removed', 'total pop']
```

```python
# evolve
history = evolve(state, deriv, timelength, **kwargs)

# visualize
print_args(**kwargs)

show(history, legend=legend, title='Initial evolution', last_t=150, **kwargs)
show(history, legend=legend, title='Full evolution', **kwargs)
```
